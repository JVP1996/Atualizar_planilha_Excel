```python
import openpyxl

def atualizar_planilha(nome, idade, cargo):
  workbook = openpyxl.load_workbook("dados.xlsx")
  aba = workbook.active
  aba.append([nome, idade, cargo])
  workbook.save("dados.xlsx")
  print("Planilha Atualizada!")

atualizar_planilha("João", 29, "Dev")
```

```python
atualizar_planilha("Clara", 26, "RH")
```
